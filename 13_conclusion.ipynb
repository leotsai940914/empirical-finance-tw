## 二｜實證投資的五大核心原則

把所有論文結論提煉成五個可以執行的原則：

---

### 原則一：承認你的大腦不適合投資（來自行為金融學）

損失厭惡、處置效應、過度自信——這些不是個人弱點，是人類的演化設定。

**解法：建立規則，取代感覺。**
- 設定停損點並嚴格執行（不靠臨場判斷）
- 定期定額自動扣款（消除「現在要不要買」的焦慮）
- 投資日誌記錄每次決策理由（讓偏誤無所遁形）

---

### 原則二：費用是唯一確定的報酬（來自三因子、0050 研究）

市場報酬不確定，但費用是確定扣走的。

| 費用差距 | 20 年後影響 |
|----------|-------------|
| 0.46% vs 1.75% | 相差約 20% 的最終資產 |
| 0.03% vs 1.75% | 相差約 28% |

**解法：費用率是選擇投資工具的第一個篩選條件，不是最後一個。**

---

### 原則三：市場因子溢酬是最穩定的（來自 CAPM + 三因子）

SMB（規模溢酬）、HML（價值溢酬）、Mom（動能溢酬）確實存在——
但它們可以連續 10 年表現不佳，且追逐它們需要更高費用和更多換手。

市場因子（Mkt-RF）最穩定，年化約 6–8%，且取得成本最低（買全市場 ETF）。

**解法：把市場因子當作核心，其他因子頂多當配角，不要本末倒置。**

---

### 原則四：時間才是最強的複利引擎（來自 DCA、Trinity、FIRE）

- 儲蓄率從 10% 提升到 30%：財務自由縮短超過 15 年
- 同樣的錢，早 10 年投入 vs 晚 10 年：最終資產可以差 1 倍以上
- 試圖擇時進出（等跌再買）：歷史上比持續投入的勝率低

**解法：「現在就開始」比「找到最佳時機」重要一千倍。**

---

### 原則五：你能忍受的波動，決定你的資產配置（來自 CAPM、Trinity）

不存在「對所有人都最好」的配置。
但 Trinity Study 告訴我們，退休提領期間，股債混合（60/40 或 80/20）
在長期歷史上比全股票或全債券都更穩健。

**解法：選一個在市場崩跌 40% 時你不會恐慌賣出的配置，然後堅持它。**

## 三｜複利的真實威力：四條人生軌跡的對比

比起討論「哪支股票好」，以下這個問題更能改變你的財富：

**「你用什麼方式，在多早的時間，把多少比例的收入投入市場？」**

下面模擬四種真實生活中都有的選擇——全部從 30 歲開始，月薪 6 萬，持續到 65 歲。

```python
import numpy as np
import matplotlib.pyplot as plt

# ── 共同參數 ────────────────────────────────────────────────
monthly_salary  = 60_000      # 月薪（元）
start_age       = 30
end_age         = 65
n_years         = end_age - start_age
months          = np.arange(0, n_years * 12 + 1)
ages            = start_age + months / 12

# ── 四條軌跡定義 ────────────────────────────────────────────
scenarios = [
    {
        'name':    '❌ 從不投資（存銀行活存）',
        'color':   '#9E9E9E',
        'monthly': monthly_salary * 0.20,   # 儲蓄率 20%
        'rate':    0.005,                    # 年利率 0.5%（台灣活存）
        'start_month': 0,
        'desc':    '儲蓄率 20%，放銀行活存 0.5%'
    },
    {
        'name':    '⚠️  主動交易（高費用、行為偏誤）',
        'color':   '#FF7043',
        'monthly': monthly_salary * 0.20,
        'rate':    0.07 - 0.02,              # 市場 7% - 行為偏誤損耗 2% - 費用 1.75% ≈ 3.25%
        'start_month': 0,
        'desc':    '儲蓄率 20%，主動選股（費用+行為偏誤損耗）實際報酬約 3.25%'
    },
    {
        'name':    '✅ 被動投資，儲蓄率 20%（標準版）',
        'color':   '#42A5F5',
        'monthly': monthly_salary * 0.20,
        'rate':    0.07 - 0.005,             # 市場 7% - 0050 費用 0.46% ≈ 6.5%
        'start_month': 0,
        'desc':    '儲蓄率 20%，低費用指數 ETF，年化 6.5%'
    },
    {
        'name':    '🚀 被動投資，儲蓄率 40%（積極版）',
        'color':   '#66BB6A',
        'monthly': monthly_salary * 0.40,
        'rate':    0.07 - 0.005,
        'start_month': 0,
        'desc':    '儲蓄率 40%，低費用指數 ETF，年化 6.5%'
    },
]

# ── 計算各軌跡的資產成長 ─────────────────────────────────────
def simulate(monthly_saving, annual_rate, n_months, start_month=0):
    monthly_rate = (1 + annual_rate) ** (1/12) - 1
    portfolio = np.zeros(n_months + 1)
    for m in range(1, n_months + 1):
        portfolio[m] = portfolio[m-1] * (1 + monthly_rate)
        if m > start_month:
            portfolio[m] += monthly_saving
    return portfolio

results = []
for sc in scenarios:
    port = simulate(sc['monthly'], sc['rate'], n_years * 12, sc['start_month'])
    results.append(port)

# ── 繪圖 ─────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 左：資產成長（百萬元）
ax = axes[0]
for sc, port in zip(scenarios, results):
    ax.plot(ages, port / 1e6, color=sc['color'], linewidth=2.5, label=sc['name'])

ax.axvline(55, color='gray', linewidth=1, linestyle=':', alpha=0.7)
ax.text(55.3, ax.get_ylim()[1] * 0.05, '55 歲\n提早退休線', fontsize=8.5, color='gray')
ax.set_xlabel('年齡', fontsize=11)
ax.set_ylabel('資產（百萬元）', fontsize=11)
ax.set_title('四條軌跡：30 歲開始，65 歲的資產差距\n（月薪 6 萬，各策略儲蓄率與報酬率不同）', fontsize=11)
ax.legend(loc='upper left', fontsize=8.5)
ax.grid(alpha=0.2)

# 右：65 歲最終資產比較（橫條圖）
ax2 = axes[1]
final_vals = [port[-1] / 1e6 for port in results]
labels_short = [
    '銀行活存\n儲蓄率20%',
    '主動交易\n儲蓄率20%',
    '被動ETF\n儲蓄率20%',
    '被動ETF\n儲蓄率40%',
]
colors = [sc['color'] for sc in scenarios]
bars = ax2.barh(labels_short, final_vals, color=colors, alpha=0.85, edgecolor='white')

for bar, val in zip(bars, final_vals):
    ax2.text(val + 0.5, bar.get_y() + bar.get_height()/2,
             f'{val:.0f}M', va='center', fontsize=11, fontweight='bold')

ax2.set_xlabel('65 歲時的資產（百萬元）', fontsize=11)
ax2.set_title('65 歲最終資產對比', fontsize=11)
ax2.grid(alpha=0.2, axis='x')

plt.tight_layout()
plt.show()

# 數字摘要
print('\n── 65 歲最終資產（月薪 6 萬，30 歲開始）──')
for sc, val in zip(scenarios, final_vals):
    total_in = sc['monthly'] * n_years * 12 / 1e6
    gain     = val - total_in
    print(f'  {sc["name"]}')
    print(f'    總投入：{total_in:.1f}M ｜ 投資增值：{gain:.1f}M ｜ 最終：{val:.0f}M')
    print()

ratio = final_vals[-1] / final_vals[0]
print(f'被動 ETF 儲蓄率 40% vs 銀行活存：相差 {ratio:.1f} 倍')
diff = final_vals[2] - final_vals[1]
print(f'被動 ETF vs 主動交易（同樣儲蓄率 20%）：差距 {diff:.1f}M')
```

## 四｜台灣投資人行動清單

這不是「理論上應該」的清單，而是結合所有學術研究後，**可以立刻執行的步驟**。

### 🔴 第一步：先把財務地基穩住（不投資前）

- [ ] **緊急備用金**：3–6 個月必要支出，放在高流動性的地方（活存、貨幣市場基金）
- [ ] **高利率負債清零**：信用卡、現金卡利率 10–15%，先還清再投資（任何投資都贏不了 15% 的資金成本）
- [ ] **計算月儲蓄**：收入 − 支出 = 每個月可以投資的金額

### 🟡 第二步：建立核心投資組合

- [ ] **選低費用的工具**：台股核心部位考慮 0050（費用率 0.46%）；若要全球布局，VT 或 AOA（費用率 0.07–0.25%）
- [ ] **設定自動扣款**：每月薪資入帳後，24 小時內自動轉入投資帳戶（消除行為偏誤）
- [ ] **決定股債比**：能承受虧損 40% 的人 → 100% 股票；無法承受者 → 60/40 或 80/20
- [ ] **不要試圖擇時**：CAPE 高時繼續投，只是降低「預期」，不是停止投入

### 🟢 第三步：保護自己不傷害自己

- [ ] **每次買賣前問自己**：我是因為「理由」在買賣，還是因為「感覺」？
- [ ] **限制看盤頻率**：每日盯盤會強化短期行為偏誤；每月檢視一次即可
- [ ] **設停損規則**：例如「單支個股超過 -20% 強制重新評估」——提前設定，不靠臨場判斷
- [ ] **對朋友的報明牌免疫**：理解「阿媽理論」——情緒高點往往是市場頂點

### 🔵 第四步：年度維護（每年做一次）

- [ ] **更新 FIRE 計算機**（Notebook 12）：檢視距離目標還有多遠
- [ ] **再平衡**：若股債比例偏離目標超過 5%，賣高買低拉回原位
- [ ] **查詢 CAPE**（multpl.com）：調整未來 10 年預期報酬的心理預期，而不是調整策略
- [ ] **檢視費用**：確認你的投資工具費用率是否有更好的替代選項

```python
# ╔══════════════════════════════════════════════╗
# ║  🔧 輸入你自己的情況，計算個人投資評分     ║
# ╚══════════════════════════════════════════════╝

my_answers = {
    '有緊急備用金（≥3個月支出）':            True,
    '高利率負債已清零':                       True,
    '有固定儲蓄並自動扣款':                   False,  # ← 改成 True 或 False
    '投資工具費用率 < 1%':                    True,
    '不會因為股市跌就停止投入':               False,
    '知道自己的 FIRE 目標金額':               False,
    '每筆交易有記錄決策理由':                 False,
    '設有明確停損/再平衡規則':                False,
    '不會因為朋友報明牌而改變計畫':           True,
    '每年至少做一次完整財務回顧':             False,
}

# ── 計算評分 ─────────────────────────────────────────────────
score = sum(1 for v in my_answers.values() if v)
total = len(my_answers)
pct   = score / total * 100

if pct >= 80:   level = ('🟢', '已建立良好投資習慣', '繼續保持，微調細節')
elif pct >= 60: level = ('🟡', '有基礎，但有明顯缺口', '聚焦在 False 的項目，一次改一個')
elif pct >= 40: level = ('🟠', '還在起步，需要調整', '優先：緊急備用金 → 自動扣款 → 低費用工具')
else:           level = ('🔴', '財務基礎需要強化', '先不要投資，把財務地基蓋穩')

emoji, rating, advice = level

# 視覺化
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# 左：儀表板
ax = axes[0]
ax.axis('off')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# 分數顯示
circle_color = '#4CAF50' if pct >= 80 else ('#FFC107' if pct >= 60 else ('#FF9800' if pct >= 40 else '#F44336'))
circle = plt.Circle((5, 6), 2.8, color=circle_color, alpha=0.15)
circle_border = plt.Circle((5, 6), 2.8, color=circle_color, fill=False, linewidth=3)
ax.add_artist(circle)
ax.add_artist(circle_border)
ax.text(5, 6.6, f'{score}/{total}', ha='center', va='center', fontsize=30,
        fontweight='bold', color=circle_color)
ax.text(5, 5.3, f'{pct:.0f} 分', ha='center', va='center', fontsize=16, color=circle_color)
ax.text(5, 9.0, '個人投資習慣評分', ha='center', fontsize=13, fontweight='bold')
ax.text(5, 4.2, f'{emoji} {rating}', ha='center', fontsize=11, color=circle_color, fontweight='bold')
ax.text(5, 3.3, advice, ha='center', fontsize=9.5, color='#555', style='italic')

# 右：各項目狀態
ax2 = axes[1]
ax2.axis('off')
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 10)
ax2.text(5, 9.6, '各項目狀態', ha='center', fontsize=12, fontweight='bold')

for i, (item, val) in enumerate(my_answers.items()):
    y = 9.0 - i * 0.82
    dot = '✅' if val else '❌'
    color = '#2E7D32' if val else '#C62828'
    ax2.text(0.3, y, dot, fontsize=10)
    ax2.text(1.0, y, item, fontsize=9, va='center', color=color)

plt.tight_layout()
plt.show()

# 文字輸出改善建議
false_items = [k for k, v in my_answers.items() if not v]
if false_items:
    print(f'\n需要改善的項目（{len(false_items)} 項）：')
    for item in false_items:
        print(f'  → {item}')
    print(f'\n建議：一次只改一個習慣，先改最容易的那個。')
```

## 五｜終身受用的心智模型

學完這門課，你需要帶走的不只是圖表和數字，而是三個**思考框架**——
面對任何投資問題，都可以用這三個問題來篩選。

---

### 框架一：「這有沒有數據？」

任何人告訴你一個投資建議，先問：
- 這個說法有沒有被實證研究驗證？
- 樣本是多大？時間跨度多長？
- 在發表之後（樣本外）還有效嗎？

> 「我去年用這個方法賺了 200%」不是數據，是故事。
> 故事的問題是，講虧損的人不會主動告訴你。

---

### 框架二：「費用和時間，是我控制得了的」

你控制不了市場走勢，你控制不了台積電今天漲不漲，
你控制不了 Fed 什麼時候降息。

但你**完全控制得了**：
- 你支付的費用率（選擇投資工具）
- 你的儲蓄率（選擇生活方式）
- 你開始的時間（現在 vs 再等等）
- 你的行為（有沒有在市場崩跌時恐慌賣出）

把注意力放在你能控制的事，忽略你控制不了的雜訊。

---

### 框架三：「這是長期規律，還是短期噪音？」

市場每天都在製造新聞。大多數新聞對長期投資者毫無意義。

一個簡單的測試：**「五年後，這件事還重要嗎？」**

- Fed 今天升息 1 碼 → 五年後？不重要。
- 台灣人口老化趨勢 → 五年後？很重要。
- 台積電昨天跌了 3% → 五年後？不重要。
- 你今天沒有執行自動扣款 → 五年後？可能很重要（複利起點）。
